```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./data/cleaned_data.secret.csv")
data
```

```python
degree_counts = data['academic_degree'].value_counts()

plt.figure()
degree_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Số lượng ý kiến của từng học vị")
plt.xlabel("Học vị")
plt.ylabel("Số lượng ý kiến")
plt.xticks(rotation=0)
plt.show()
```

```python
faculty_counts = data['faculty'].value_counts()

plt.figure()
faculty_counts.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title("Number of Comments per Faculty")
plt.xlabel("Faculty")
plt.ylabel("Number of Comments")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
criteria_counts = data['criteria'].value_counts().head(15)

plt.figure(figsize=(10, 6))
criteria_counts.plot(kind='barh', color='salmon', edgecolor='black')
plt.title("Top 15 Criteria by Number of Comments")
plt.xlabel("Number of Comments")
plt.ylabel("Criteria")
plt.gca().invert_yaxis()  # Show the largest at the top
plt.show()
```

```python
result = data[data['criteria'] == 'Phục vụ']
print(result)
```

```python
comment_per_user = data['comment_id'].value_counts()

plt.figure()
plt.hist(comment_per_user, bins=range(1, comment_per_user.max() + 2), 
         color='mediumpurple', edgecolor='black', alpha=0.8)
plt.title("Histogram: Number of Comments per User")
plt.xlabel("Number of Comments per User")
plt.ylabel("Number of Users")
plt.show()
```
