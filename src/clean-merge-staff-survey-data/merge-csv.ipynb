```python
import pandas as pd
```

```python
api_csv = pd.read_csv('./data/api-staff-survey-data.secret.csv')
xlsx_csv = pd.read_csv('./data/xlsx-staff-survey-data.secret.csv')
```

```python
api_csv['comment'] = api_csv['comment'].astype(str).str.strip()
xlsx_csv['comment'] = xlsx_csv['comment'].astype(str).str.strip()
```

```python
new_rows = api_csv[~api_csv['comment'].isin(xlsx_csv['comment'])].copy()
for col in ['faculty', 'academic_degree']:
    if col not in new_rows.columns:
        new_rows[col] = ''

print(f"Total new rows: {len(new_rows)}")
new_rows
```

```python
xlsx_csv_updated = pd.concat([xlsx_csv, new_rows], ignore_index=True)
```

```python
xlsx_csv_updated
```

```python
xlsx_csv_updated.to_csv('./data/all-staff-survey-data.secret.csv', index=False)
```
